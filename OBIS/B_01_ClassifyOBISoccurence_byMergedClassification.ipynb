```python
import pandas as pd
import os
```

```python
df_OBIS_dataset_classification = pd.read_csv('Dataset_and_Org_classifications/OBIS_dataset_metrics.csv')
df_OBIS_dataset_classification.set_index('dataset_id', inplace=True)
```

```python
df_OBIS_dataset_classification
```

```python
# Function to process chunks
def process_chunk(chunk, df_OBIS_dataset_classification):
    # Merge the chunk with df_OBIS_dataset_classification to add PrivateFlag
    chunk = chunk.merge(df_OBIS_dataset_classification[['PrivateFlag']], left_on='dataset_id', right_index=True, how='left')
    return chunk

```

```python
input_OBIS_occurence_file = 'input_occurence_data/obis_filtered.csv'

# Initialize progress tracking
chunk_size = 10**6  # Adjust chunk size as needed
total_rows = 128621187  # Hardcoded total number of rows
total_chunks = (total_rows // chunk_size) + 1
```

```python
output_path = 'output_occurence_data_classified/obis_classified.csv'
private_output_path = 'output_occurence_data_classified/obis_classified_privateonly.csv'
```

```python
# # Check loading of the file (and if necessary QC length)
# dfOBIS = pd.read_csv(input_OBIS_occurence_file, nrows=10000)
# dfOBIS
```

```python
# Check if output files already exist
if os.path.exists(output_path) or os.path.exists(private_output_path):
    print("One or both of the output files already exist.")
    print("Please delete or rename the following files before proceeding:")
    if os.path.exists(output_path):
        print(f"- {output_path}")
    if os.path.exists(private_output_path):
        print(f"- {private_output_path}")
    print("Aborting the process.")
    
else:
    # Write headers to output files
    header_written = False
    private_header_written = False

    # Open files for writing
    with open(output_path, 'w') as all_file, open(private_output_path, 'w') as private_file:
        # Process dfOBIS in chunks
        for i, chunk in enumerate(pd.read_csv(input_OBIS_occurence_file, chunksize=chunk_size)):
            # Process the chunk
            processed_chunk = process_chunk(chunk, df_OBIS_dataset_classification)
            
            # Write processed chunk to the main CSV file
            processed_chunk.to_csv(all_file, mode='a', header=not header_written, index=False)
            
            # Write rows where PrivateFlag is True to the private CSV file
            private_rows = processed_chunk[processed_chunk['PrivateFlag'] == True]
            private_rows.to_csv(private_file, mode='a', header=not private_header_written, index=False)
            
            # Update progress
            print(f'\rProcessing: Chunk {i + 1}/{total_chunks}', end='')
            
            # Update header flags
            header_written = True
            private_header_written = True

    # Print final message
    print(f'\nProcessing complete. Outputs written to {output_path} and {private_output_path}')
```
